```python
import json
import os
import numpy as np
import pandas as pd
```

```python
path= '/Users/debodeepbanerjee/Downloads/clevr/'
im_path = path+'images/'
cap_path=path+'scenes'
```

```python
with open(f'{cap_path}/CLEVR_train_scenes.json', 'r') as file:
    trn_scenes=json.load(file)

with open(f'{cap_path}/CLEVR_val_scenes.json', 'r') as file:
    val_scenes=json.load(file)

with open(f'{cap_path}/CLEVR_test_scenes.json', 'r') as file:
    test_scenes=json.load(file)
```

```python
def generate_captions(file):
    sentences = []
    for i in range(len(file['scenes'])):
        objects = file['scenes'][i]['objects']
        relations = file['scenes'][i]['relationships']
        captions = ''
        for j in range(len(objects)):
            temp_captions = f"{objects[j]['size']} {objects[j]['color']} {objects[j]['material']} {objects[j]['shape']} "
            relation_cap = "behind "
            if len(relations['front'][j]) > 0:
                # print('yes')
                _ = ''
                for k in relations['front'][j]:
                    temp_rel = f"{objects[k]['color']} {objects[k]['shape']}, "
                    _ = _+temp_rel
            else:
                _ = 'nothing '
            relation_cap = relation_cap+_+";"
            temp_captions = temp_captions+relation_cap
            # print(temp_captions+relation_cap)
            captions += temp_captions
        sentences.append(captions)
    return sentences       
```

```python
train_sentences = generate_captions(trn_scenes)

count = 0
for i in train_sentences:
    count += len(i.split())
avg_len = count/len(train_sentences)
print(avg_len)

val_sentences = generate_captions(val_scenes)

count = 0
for i in val_sentences:
    # print(len(i.split()))
    count += len(i.split())
avg_len = count/len(val_sentences)
print(avg_len)

test_sentences = generate_captions(test_scenes)

count = 0
for i in test_sentences:
    count += len(i.split())
avg_len = count/len(test_sentences)
print(avg_len)
```

```python
# As the image ids are sequential, it is easier to assign the ids and paths.
train_image_id=[]
train_image_paths=[]
for i in range(len(train_sentences)):
    image_id = f"CLEVR_train_{i:06d}"
    image_path = f'/home/debodeep.banerjee/synthetic/clevr_images/train/{image_id}.png'
    train_image_id.append(image_id)
    train_image_paths.append(image_path)

train_data = pd.DataFrame({'image_id':train_image_id, 'image_path':train_image_paths, 'captions':train_sentences})
# train_data['split']='train'
train_data.head()
```
